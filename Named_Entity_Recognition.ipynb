```python
import spacy

nlp=spacy.load("en_core_web_sm")

#sentence="Tesla Inc is going to acquire Twitter Inc for $45 billion"
sentence="Apple is looking at buying U.K. startup for $1 billion."
```

```python
#first way
doc=nlp(sentence)

for ent in doc.ents:
    print(ent.text,"->",ent.label_,"->",spacy.explain(ent.label_))
```

```python
#second way
from spacy import displacy

displacy.render(doc,style="ent")
```

```python
sent="Tesla is going to acquire Twitter for $45 billion"
doc1=nlp(sent)
print(doc1[2:4])
print(type(doc1[2:4]))
```

```python
from spacy.tokens import Span

s1=Span(doc1,0,1,label="ORG")
s2=Span(doc1,5,6,label="ORG")

doc1.set_ents([s1,s2],default="unmodified")
```

```python
for ent in doc1.ents:
    print(ent.text,"->",ent.label_,"->",spacy.explain(ent.label_))
```
