```python
import pandas as pd
data={
    "text":["people watch youtube","youtube watch youtube","people write comment","youtube write comment"],
    "num":[10,20,30,40],
    
}
df=pd.DataFrame(data)
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(ngram_range=(2,2))
#(1,1) bag of words
#(1,2) bi gram,Uni gram
#(2,2) bi gram

count_vectorizer=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(count_vectorizer[0].toarray())
print(count_vectorizer[1].toarray())
print(count_vectorizer[2].toarray())
print(count_vectorizer[3].toarray())
```
